```python
for run in runs:
    outputFolderZZ = f'./outV2/theta{pfcParmsZZ.itheta:03d}/noise05/run{runs[0]}'
    if outputFolderZZ is not None:
        os.makedirs(outputFolderZZ, exist_ok=True)

    simZZ = pfc.pfc2d()
    simZZ.initParms(pfcParmsZZ)

    runParms = [
        {'dt': 1e-5, 'steps': 10000, 'printEvery': 1000, 'saveEvery': 5000, 'outputFolder': outputFolderZZ,},
        {'dt': 1e-4, 'steps': 9000, 'printEvery': 1000, 'saveEvery': 4500, 'outputFolder': outputFolderZZ,},
        {'dt': 1e-3, 'steps': 9000, 'printEvery': 1000, 'saveEvery': 4500, 'outputFolder': outputFolderZZ,},
        {'dt': 1e-2, 'steps': 9000, 'printEvery': 1000, 'saveEvery': 4500, 'outputFolder': outputFolderZZ,},
        {'dt': 1e-2, 'steps': 90000, 'printEvery': 10000, 'saveEvery': 45000, 'outputFolder': outputFolderZZ,},
        ]

    for runParm in runParms:
        simZZ.run(**runParm)
```
